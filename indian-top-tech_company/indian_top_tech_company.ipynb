```python
import matplotlib.pyplot as plt
import numpy as np

company = ['TCS','HCL','Wipro','Infosys','Tech-Mahindra']
comp = ['Google','Tesla','Microsoft','IBM','Meta']
revenue = [120,300,90,130,80]
profit = [60,140,40,60,30]
sold = [40,70,20,30,10]
total = [100,180,80,110,60]
lamao = [200,400,250,350,150]
ypos = np.arange(len(company))
plt.xticks(ypos,company)
plt.title('Indian Top Tech Companies')
plt.ylabel('Revenue(bln)')
plt.bar(ypos-0.08,revenue,label='revenue',width=0.16)
plt.bar(ypos+0.08,profit,label='profit',color='yellow',width=0.16)
plt.bar(ypos+0.24,sold,label='sold',color='red',width=0.16)
plt.bar(ypos+0.40,total,label='total',color='green',width=0.16)
plt.bar(ypos+0.56,lamao,label='lamao',color='pink',width=0.16)
plt.legend()
plt.grid()
plt.show()
```
